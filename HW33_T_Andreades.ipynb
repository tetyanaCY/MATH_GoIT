# **Завдання 1**

Якщо розглянути більярдний стіл як площину, то швидкість більярдної кулі до удару в борт була (3,2). Удар в борт надав кулі швидкість, яку можна описати вектором (0,-4). Яка швидкість у кулі після удару? Сформуй графічно та із допомогою бібліотеки NumPy обрахунки для даної задачі.

Порахуй кут, під яким вдарилася куля.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Вектори
v_ball_before = np.array([3, 2])  # Швидкість кулі до удару
v_impact = np.array([0, -4])      # Швидкість удару

# Розрахунок швидкості кулі після удару
v_ball_after = v_ball_before + v_impact

# Розрахунок кута між векторами
angle_radians = np.arccos(np.dot(v_ball_before, v_impact) / (np.linalg.norm(v_ball_before) * np.linalg.norm(v_impact)))
angle_degrees = np.degrees(angle_radians)

# Візуалізація
plt.figure(figsize=(8, 6))

# Зафарбування столу
plt.fill_between([-5, 5], -2, 2, color='green', alpha=0.3)

# Візуалізація векторів
plt.quiver(0, 0, v_ball_before[0], v_ball_before[1], angles='xy', scale_units='xy', scale=1, color='r', label='Швидкість до удару')
plt.quiver(0, 0, v_impact[0], v_impact[1], angles='xy', scale_units='xy', scale=1, color='b', label='Швидкість удару')
plt.quiver(0, 0, v_ball_after[0], v_ball_after[1], angles='xy', scale_units='xy', scale=1, color='orange', label='Швидкість після удару')
plt.quiver(v_ball_before[0], v_ball_before[1], v_impact[0], v_impact[1], angles='xy', scale_units='xy', scale=1, color='gray', label='Зміна швидкості внаслідок удару')

# Налаштування графіка
plt.xlim(-6, 6)
plt.ylim(-3, 3)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.title('Вектори швидкості більярдної кулі на столі')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Виведення результату обчислення кута
plt.figtext(0.5, 0.01, f"Кут удару: {angle_degrees:.2f}°", ha="center", fontsize=12, color='purple')

plt.show()

```

**Основні елементи графіка:**

*  *Червона стрілка ("Швидкість до удару"): вектор швидкості кулі до удару*
(3,2).
* *Синя стрілка ("Швидкість удару"): вектор швидкості удару*
(0,−4).
* *Помаранчева стрілка ("Швидкість після удару"): вектор швидкості кулі після удару* (3,−2).
*  *Сіра стрілка ("Зміна швидкості внаслідок удару"): показує зміну швидкості в результаті удару.*
*   Кут удару, який становить 123.69 градусів, вказано внизу графіка (кут між синім та червоними векторами).
*   Більярдний стіл зображено зеленим кольором

**Завдання 2**

Дві кулі однакової маси на більярді вдарилися одна об одну. Перша куля до удара котилася зі швидкістю, що можна описати вектором (0,5). А після удару перша куля катиться зі швидкістю (1,2), а друга повністю зупинилась. Будемо вважати, що при ударі імпульс не було втрачено (тобто виконується рівність m1v1⃗+m2u1⃗=m1v2⃗+m2u2⃗, де m1,m2 - маси куль, v1⃗,u1⃗ - вектори швидкостей куль до удару, v2⃗,u2⃗ - вектори швидкостей куль після удару). Який вектор швидкості другої кулі до удару?


```python
import matplotlib.pyplot as plt
import numpy as np


# Вектори швидкості
v1_initial = np.array([0, 5])  # Початкова швидкість першої кулі
v1_final = np.array([1, 2])    # Кінцева швидкість першої кулі
u1_final = np.array([0, 0])    # Швидкість другої кулі після зіткнення (зупинилася)

# Знаходимо вектор швидкості другої кулі до зіткнення u1
# u1 = v1_final + u1_final - v1_initial
u1_initial = v1_final + u1_final - v1_initial

# Побудова графіку з клітинками 1x1
plt.figure(figsize=(10, 6))
plt.grid(which='both', linestyle='-.', linewidth=0.5)

# Траєкторії та вектори швидкості до зіткнення
plt.quiver(0, 0, v1_initial[0], v1_initial[1], angles='xy', scale_units='xy', scale=1, color='blue', label="Початкова швидкість кулі 1 (0,5)")
plt.quiver(0, 0, u1_initial[0], u1_initial[1], angles='xy', scale_units='xy', scale=1, color='red', label="Початкова швидкість кулі 2 (1,-3)")

# Траєкторії та вектори швидкості після зіткнення
plt.quiver(0, 0, v1_final[0], v1_final[1], angles='xy', scale_units='xy', scale=1, color='green', label="Кінцева швидкість кулі 1 (1,2)")
plt.quiver(0, 0, u1_final[0], u1_final[1], angles='xy', scale_units='xy', scale=1, color='orange', label="Кінцева швидкість кулі 2 (0,0)")

# Налаштування масштабу осей
plt.xticks(np.arange(-5, 6, 1))
plt.yticks(np.arange(-5, 6, 1))

# Додавання деталей до графіку
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.title('Зіткнення більярдних куль: Вектори швидкості до і після зіткнення')
plt.xlabel('Компонента X')
plt.ylabel('Компонента Y')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')

# Відображення графіку
plt.show()

# Вивід обчисленого вектора швидкості другої кулі до зіткнення
u1_initial
```

**Завдання 3**

Товари в галузі виробляються зі швидкістю, яку можна описати вектором (1,5,2,6,7). Через негативні форс-мажорні явища зміни за поточний рік можна охарактеризувати як (0,-1,0,-0,-1), проте в галузі зберігається зростання на рівні 20% на рік. За який час галузь повністю подолає наслідки форс-мажорних обставин?

```python
import numpy as np

# Задаємо вектори
a = np.array([1, 5, 2, 6, 7])  # вектор швидкості виробництва
b = np.array([0, -1, 0, 0, -1]) # вектор змін через форс-мажорні обставини

# Перевіряємо, чи всі елементи вектора швидкості додатні
all_positive = np.all(a > 0)

# Знаходимо вектор темпу зростання
growth_rate = 0.2
da = a * growth_rate

# Задаємо вектор зміни
change = np.copy(b)
years = 0

# Конвертуємо вектор зміни у float для сумісності з вектором темпу зростання
change = b.astype(float)

# Перераховуємо час, необхідний для подолання наслідків форс-мажорних обставин
years = 0
while not np.all(change - (-b) >= 0):
    change += da
    years += 1

all_positive, years
# Виводимо результат обчислень
result = f"Галузі знадобиться {years} роки, щоб повністю подолати наслідки форс-мажорних обставин."
result
```

Тут можу посперечатися з логікою завдання, бо немає ясності щодо зростання в галузі, з якого самого моменту.
**Якщо з початку поточного року, тоді галузі знадобиться 1 рік.**

Завдання полягає у визначенні, за який час вплив форс-мажорних подій буде повністю нейтралізовано завдяки зростанню в галузі. Тут ми можемо розглянути кожен вид товару окремо, оскільки зменшення відбулося лише у двох видах товарів.

**Для другого виду товару:** початковий обсяг виробництва 5 одиниць, зменшення на 1 одиницю, щорічне зростання на 20%.(5*20%=1 одиниця - зростання)

**Для п'ятого виду товару:** початковий обсяг виробництва 7 одиниць, зменшення на 1 одиницю, щорічне зростання на 20%. (7*20%=1,4 одиниця - зростання)

Тепер  обрахуваємо час, необхідний для подолання наслідків форс-мажорних обставин для кожного з цих двох видів товарів, і оберемо **найбільший** із отриманих результатів. (1 рік для двох видів товарів).Негативні наслідки - зменшення по одній одиниці за рік по двум групам товарів.

Галузь повністю подолає наслідки форс-мажорних обставин за 1 рік. Це означає, що зростання в галузі на рівні 20% на рік дозволить нейтралізувати вплив негативних подій на виробництво за цей період.
