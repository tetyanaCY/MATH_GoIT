```python
import numpy as np

# Введення  даних з гугл драйва https://docs.google.com/spreadsheets/d/1o9nWslMKiLFt3BW5avN-XYuXkNISTpXK1WtY9M4vTQg/edit#gid=638659925
# Вектор А - мої оцінки життєвого балансу
A = np.array([7, 9, 6, 6, 9, 5, 5, 5])  # "Кар'єра", "Фінанси", "Оточення", "Розвиток (освіта), "Відпочинок", "Подорожі", "Здоров'я, спорт"

# Вектор В - кількість годин на тиждень, відведених на кожну сферу
B = np.array([40, 16, 1, 3, 15, 7, 20, 14])

# Обрахунок суми елементів вектора А (поточний показник щастя)
sum_A = np.sum(A)

# Обрахунок вектора С (ефективність)
C = A / B

# Новий розподіл часу - вектор D
D = np.array([40, 20, 2, 3, 12, 9, 12, 18])

# Обрахунок вектора E (прогнозоване задоволення)
E = D * C

# Обрахунок суми елементів вектора E (прогнозований показник щастя)
sum_E = np.sum(E)

# Виведення результатів
print("Поточний показник щастя:", sum_A)
print("Ефективність за кожною сферою (вектор C):", C)
print("Прогнозоване задоволення від кожної сфери (вектор E):", E)
print("Прогнозований показник щастя:", sum_E)










```

```python
import numpy as np
import pandas as pd

# Введення даних
A = np.array([7, 9, 6, 6, 9, 5, 5, 5])
B = np.array([40, 16, 1, 3, 15, 7, 20, 14])
D = np.array([40, 20, 2, 3, 12, 9, 12, 18])

# Обрахунки
C = A / B
E = D * C
sum_A = np.sum(A)
sum_E = np.sum(E)

# Створення DataFrame
df = pd.DataFrame({
    "Сфера": ["Кар'єра", "Сім'я", "Фінанси", "Оточення", "Освіта", "Відпочинок", "Подорожі", "Здоров'я, спорт"],
    "Оцінка (A)": A,
    "Години (B)": B,
    "Ефективність (C)": C,
    "Перерозподіл годин (D)": D,
    "Прогнозоване задоволення (E)": E
})

# Виведення таблиці
print("Поточний показник щастя:", sum_A)
print("Прогнозований показник щастя:", sum_E)
print("\nДетальний аналіз:")
print(df)
```
